```python
weather.drop(columns='wind_gust',inplace=True)#only 70k entries and a gust is by definition temporary and has negligible effect on delays
#since we have plenty of data, we can simply drop rows with na values. Other option would be to use the mean value of that column and fill the na values. Choose to drop.
weather.dropna(subset=['temperature','dewpoint','humidity','wind_direction','wind_speed','precipitation','pressure','visibility'],inplace=True)
weather.info()
```

```python
#print(np.round(1.29-0.3+0.5)) #used to check rounding threshold trick
```

```python
#join the weather data into the flight data based on the weather @ flight departure time @ right airport
decrease_round_threshold_to=0.3 #so that 5.30pm-6.30pm flights use the 6pm weather data 
flights['hour']=np.round((flights['departure']/100)-decrease_round_threshold_to+0.5).astype(int) #assume weather is same at arrival as at departure
flights_weather=pd.merge(flights,weather,how='inner',on=['month','day','hour','year','origin'])
flights_weather.info() #data free of na values
```

```python
#replace arrival entries of 2400 with 0 --> timedata only accepts 0-2359
flights_weather[flights_weather['arrival']==0]
arrivals=flights_weather['arrival']+0
flights_weather['arrival']=np.where(arrivals==2400,0,arrivals)

#convert data into usable time format for departure, scheduled departure, arrival and scheduled_arrival
getTime = lambda t: t[:2] + ':' + t[2:]

flights_weather['departure']=flights_weather['departure'].astype(int).astype(str).str.zfill(4).apply(getTime)
flights_weather['departure']=pd.to_datetime(flights_weather['departure'], format='%H:%M').dt.time
flights_weather['departure']=pd.to_timedelta(flights_weather['departure'].astype(str))

flights_weather['scheduled_departure']=flights_weather['scheduled_departure'].astype(int).astype(str).str.zfill(4).apply(getTime)
flights_weather['scheduled_departure']=pd.to_datetime(flights_weather['scheduled_departure'], format='%H:%M').dt.time
flights_weather['scheduled_departure']=pd.to_timedelta(flights_weather['scheduled_departure'].astype(str))

flights_weather['arrival']=flights_weather['arrival'].astype(int).astype(str).str.zfill(4).apply(getTime)
flights_weather['arrival']=pd.to_datetime(flights_weather['arrival'], format='%H:%M').dt.time
flights_weather['arrival']=pd.to_timedelta(flights_weather['arrival'].astype(str))

flights_weather['scheduled_arrival']=flights_weather['scheduled_arrival'].astype(int).astype(str).str.zfill(4).apply(getTime)
flights_weather['scheduled_arrival']=pd.to_datetime(flights_weather['scheduled_arrival'], format='%H:%M').dt.time
flights_weather['scheduled_arrival']=pd.to_timedelta(flights_weather['scheduled_arrival'].astype(str))
```

```python
#calculating flight delays

import datetime
#consider overnight case: departure = 1 and scheduled departure = 2359
delay_dep=flights_weather['departure']-flights_weather['scheduled_departure'] #delay=dep-sch
delay_dep=np.where(delay_dep<datetime.timedelta(minutes=0),0,delay_dep) #only considers delays, not early departures
delay_dep=pd.Series(delay_dep)

delay_arr=flights_weather['arrival']-flights_weather['scheduled_arrival']
delay_arr=np.where(delay_arr<datetime.timedelta(minutes=0),0,delay_dep)
delay_arr=pd.Series(delay_arr)
```

```python
#merging the delay calculations with the weather_flight data
flights_weather=pd.concat([flights_weather,delay_arr,delay_dep],axis=1) #merge new columns onto df
flights_weather.rename(columns={0:'delay_arr',1:'delay_dep','distance':'Distance (miles)'},inplace=True) #rename concatenated columns
flights_weather['Delay (min)']=flights_weather['delay_arr']+flights_weather['delay_dep'] #add total delay column
flights_weather['Delay (float)']=flights_weather['Delay (min)'].dt.total_seconds()/60 #create delay of floats for later regression
flights_weather['Departure (float)']=flights_weather['departure'].dt.total_seconds()
flights_weather['Arrival (float)']=flights_weather['arrival'].dt.total_seconds()

#data is now clean
#EWR <-> ERW could still be made equal should more data be needed and relevant assumption made
```

```python
#Q2) Examining the relationship between distance and delay
sns.set_palette("coolwarm")
sns.set_style('white')
fig=sns.jointplot(flights_weather['Distance (miles)'],flights_weather['Delay (min)'].astype('timedelta64[m]'))
plt.suptitle('A graph showing the distribution and correlation of flight delays over different distance flights\n')
#delay seems to be pretty uncorrelated to distance
```
