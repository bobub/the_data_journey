# Predicting Employment

Predict EMPLOYED/SEEKING EMPLOYMENT based on sex, population group, intl or SA or home country, level of study or academic career code, awarded degree code or academic program code

```python
df2=df[(df['MAINACTIVITY']=='EMPLOYED')|(df['MAINACTIVITY']=='SEEKING EMPLOYMENT')] #only look at employed or seeking employment, not those studying, seeking studies or gap years.
df2.info()
```

```python
df2=df2[['MAINACTIVITY','StudentData.Sex', 'StudentData.Population Group', 'StudentData.International Or SA','StudentData.History Level Of Study', 'StudentData.Awarded Degree Code']]
df2.dropna(inplace=True)
```

```python
#create dummy variables for numerous categorical features
cat_feats = ['StudentData.Sex', 'StudentData.Population Group', 'StudentData.International Or SA','StudentData.History Level Of Study', 'StudentData.Awarded Degree Code']
final_data = pd.get_dummies(df2,columns=cat_feats,drop_first=True)
```

```python
final_data.info()
```

```python
#train test split
from sklearn.model_selection import train_test_split
X = X = final_data.drop('MAINACTIVITY',axis=1)
y = final_data['MAINACTIVITY']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#logistic regression model
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions_log_model = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions_log_model))
```
